In the first task you should write code in which you:

 - Read following datasets from the OSM:

 - Select the buildings and roads that intersect with the given administrative boundary (district)

````python
# Import the necessary libraries
import osmnx as ox
import matplotlib.pyplot as plt

# Set up the plotting style
plt.style.use("dark_background")

# Define the place
place_name = "Otaniemi Aalto, Espoo, Finland"
place_address= """Otaniemi, Suur-Tapiola, Espoo, Helsinki sub-region,
Uusimaa, Southern Finland, Mainland Finland,
02150, Finland"""

graph = ox.graph_from_address(place_address)

# Fetch OSM street network from the location
# graph = ox.graph_from_place(place_name, network_type='drive')

# Fetch buildings from the location
# buildings = ox.geometries_from_place(place_name, tags={'building': True})
buildings = ox.geometries_from_address(place_address, tags={'building': True})

# Fetch the boundary of the administrative region
boundary = ox.geocode_to_gdf(place_address)

# Project the street network to UTM
graph_projected = ox.project_graph(graph)

# Get the projected graph's crs
graph_crs = ox.graph_to_gdfs(graph_projected, nodes=False).crs

# Project buildings and boundary to the same CRS as the projected graph
buildings_projected = buildings.to_crs(graph_crs)
boundary_projected = boundary.to_crs(graph_crs)

# Select the buildings and roads that intersect with the given administrative boundary (district)
boundary_polygon = boundary_projected.unary_union
buildings_in_boundary = buildings_projected[buildings_projected.intersects(boundary_polygon)]
roads_in_boundary = ox.graph_to_gdfs(graph_projected, nodes=False, edges=True)[ox.graph_to_gdfs(graph_projected, nodes=False, edges=True).intersects(boundary_polygon)]

# Reproject the selected buildings and roads to EPSG:3067 (ETRS-TM35FIN)
buildings_in_boundary = buildings_in_boundary.to_crs(epsg=3067)
roads_in_boundary = roads_in_boundary.to_crs(epsg=3067)
boundary_projected = boundary_projected.to_crs(epsg=3067)

# Start plotting
fig, ax = plt.subplots(figsize=(12, 12))

# Plot buildings
buildings_in_boundary.plot(ax=ax, color='white', alpha=0.7)

# Plot roads
roads_in_boundary.plot(ax=ax, linewidth=1, edgecolor='white')

# Plot boundary
boundary_projected.boundary.plot(ax=ax, color='red', linewidth=2)

# Add a title to the plot
plt.title('Map of Otaniemi', fontsize=15, color='white')

# Remove axis for aesthetics
# plt.axis('off')

# Show the plot
# plt.show()
plt
````

````python
def plot_Ripleys_test(test, 
                      coordinates,
                      title="", 
                      xlabel="distance", 
                      ylabel='% of nearest neighbor\ndistances shorter',
                      xlim=None,
                      ):
    """
    Helper function to plot results from Ripley's tests. 
    Modified after Rey, Arribas-Bel & Wolf (2021). Geographic Data Science with Python.
    """
    
    import matplotlib.pyplot as plt
    import numpy as np

    f,ax = plt.subplots(1,2,figsize=(9,3), 
                        gridspec_kw=dict(width_ratios=(6,3)))
    
    # Plot the simulations with thin lines
    ax[0].plot(test.support, test.simulations.T, color='k', alpha=.01)
    
    # Show the average of simulations
    ax[0].plot(test.support, np.median(test.simulations, axis=0), color='cyan', 
             label='median simulation (CSR)')


    # Show the observed pattern's G|F|K function
    ax[0].plot(test.support, test.statistic, label = 'observed', color='red')

    # Clean labels and axes
    ax[0].set_xlabel(xlabel)
    ax[0].set_ylabel(ylabel)
    ax[0].legend()
    
    # If xlim is provided modify accordingly
    if xlim is not None:
        ax[0].set_xlim(0, xlim)
    else:
        ax[0].set_xlim(0, round(test.support.max(), 0))
        
    # Set title for the plot
    ax[0].set_title(title)

    # Plot the point pattern 
    ax[1].scatter(*coordinates.T)

    # Clean up labels and axes from the point pattern
    ax[1].set_xticks([])
    ax[1].set_yticks([])
    ax[1].set_xticklabels([])
    ax[1].set_yticklabels([])
    ax[1].set_title('Pattern')
    f.tight_layout()
    plt.show()


# REPLACE THE ERROR BELOW WITH YOUR OWN CODE
raise NotImplementedError()
````
